```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('KaggleV2-May-2016.csv')
df
```

```python
df.columns = df.columns.str.strip().str.lower()
df
```

```python
# pd.to_datetime(df[['scheduledday','appointmentday' ]], errors = 'coerce')    gives error because to datetime function is appliciable only on series type

df['scheduledday'] = pd.to_datetime(df['scheduledday'], errors='coerce')
df['appointmentday'] = pd.to_datetime(df['appointmentday'], errors='coerce')
df
```

```python
df['gender'].unique()
```

```python
df['age'].unique()
```

```python
df = df[(df['age'] >= 0) & (df['age'] <= 100)]
df['age'].unique()
```

```python
df['scholarship'].value_counts()
```

```python
df[['scholarship', 'hipertension', 'diabetes', 'alcoholism','handcap', 'sms_received' ]] = df[['scholarship', 'hipertension', 'diabetes', 'alcoholism','handcap', 'sms_received' ]].replace({0:'no', 1:'yes'})
df
```

```python
df['appointment_delay'] = (df['scheduledday'] - df['appointmentday']).dt.days
df
```

```python
df['appointment_delay'] = (df['appointmentday'] - df['scheduledday']).dt.days

col = df.pop('appointment_delay')
df.insert(5, 'appointment_delay', col)
df
```

```python
df = df[df['appointment_delay']>=0]
df
```

```python
df.rename(columns={'No-show': 'appointment_status'}, inplace=True)
df
```

```python
sexrate = df['gender'].value_counts()
sexrate
```

```python
# Females care more of thier health

plt.pie(sexrate.values, labels = sexrate.index, autopct = '%1.1f%%')
plt.title('Sex Rate')
plt.show()
```

```python
# Month is which maximum appointments are scheduled
df['scheduledday'].dt.month.value_counts().sort_index().plot(kind='bar')
```

```python
sns.boxplot(x='no-show', y='appointment_delay', data=df)
plt.title('Gap between Scheduled Day and Appointment Day by No-show')
plt.show()
```

```python
bins = [0, 12, 18, 30, 45, 60, 100]
labels = ['0-12','13-18','19-30','31-45','46-60','60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Countplot
sns.countplot(x='age_group', hue='no-show', data=df)
plt.title('No-show by Age Group')
plt.show()

# Or percentage
age_noshow = df.groupby('age_group')['no-show'].value_counts(normalize=True).unstack()
print(age_noshow)
```
